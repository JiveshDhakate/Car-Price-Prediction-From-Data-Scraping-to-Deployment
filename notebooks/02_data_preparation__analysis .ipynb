# **Data Preprocessing**

## **Standardizing Dates**

The **"Date of Sale"** column contains multiple date formats, including:
- **`DD/MM/YYYY`** (e.g., `06/01/2024`)
- **`YYYY-MM-DD`** (e.g., `2024-01-09`)

To ensure consistency, we standardized all dates to **`YYYY-MM-DD` format** using `date_standardizer()`.

---

### **Key Steps in Date Processing**
1. **Identified unique date formats** in the dataset.
2. **Converted all dates to a uniform format** (`YYYY-MM-DD`).
3. **Checked for incorrect values**, ensuring no `"None"` values remain.

---

### **Results**
- The dataset now has a **consistent date format**.
- **Missing or unrecognized dates remain as `None`** without altering data integrity.
- **Validation check:**  
  ```python
  (cars_df['Date of Sale'] == 'None').sum()



````python
unique_dates = cars_df['Date of Sale'].dropna().unique()
print('Date Formats Present in Merged Dataset\n',unique_dates[:50])
````

````python
import pandas as pd
import re

def date_standardizer(df, date_column):
    def parse_date(date):
        if pd.isna(date) or date == '' or date == 'NaN':
            return None

        date = str(date).strip()

        if re.match(r'^\d{4}-\d{2}-\d{2}$', date):
            return pd.to_datetime(date, format='%Y-%m-%d', errors='coerce')

        elif re.match(r'^\d{2}/\d{2}/\d{4}$', date):
            return pd.to_datetime(date, format='%d/%m/%Y', errors='coerce')

        else:
            return None

    df[date_column] = df[date_column].apply(parse_date)

    df[date_column] = df[date_column].dt.strftime('%Y-%m-%d')

    return df

cars_df = date_standardizer(cars_df, 'Date of Sale')


value=(cars_df['Date of Sale']=='None').sum()
print('Number of Dates with None Value: ',value)
````

## **Creating 'Car Age at Sale' Feature**
The "Car Age at Sale" feature was created to quantify the number of years between a car’s manufacturing year and the date it was sold. This provides a clearer understanding of vehicle depreciation over time.

---

**Key Steps in Car Age Processing**
- Ensured the "Date of Sale" column was in proper `datetime` format using `pd.to_datetime()`.
- Extracted the year from the "Date of Sale" column.
- Subtracted the "Year" (of manufacture) from the extracted sale year.
- Created a new column named `"Car Age at Sale"` to store the result.

---

**Results**
- The dataset now includes a new feature that captures how old a car was at the time of sale.
- This provides a more meaningful representation of depreciation.
- Helps the machine learning model understand the relationship between car age and resale price.
- Enhances the overall predictive quality of the dataset.


````python
# Ensure Date of Sale is in datetime format
cars_df['Date of Sale'] = pd.to_datetime(cars_df['Date of Sale'], errors='coerce')

# Extract year from Date of Sale and subtract Year of Manufacture
cars_df['Car Age at Sale'] = cars_df['Date of Sale'].dt.year - cars_df['Year']

# Preview
cars_df.head()
````
